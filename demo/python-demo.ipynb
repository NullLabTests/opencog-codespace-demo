# OpenCog AtomSpace — Python Demos

Connect to a running CogServer from Python and interact with the AtomSpace.

**Prerequisites:** A CogServer running on `localhost:18080`.

```bash
pip install websockets requests jupyter
```

````python
import asyncio
import json
import websockets

async def ask(expr, host="localhost", port=18080):
    """Send a Scheme expression to the CogServer and return the result."""
    uri = f"ws://{host}:{port}/json"
    async with websockets.connect(uri) as ws:
        await ws.send(json.dumps({"command": "scheme", "body": expr}))
        resp = await ws.recv()
        return json.loads(resp)
````

````python
async def main():
    # Create atoms
    r = await ask('(Concept "jupyter-demo")')
    print("Created:", r.get("result", r))

    # Build inheritance
    r = await ask('(Inheritance (Concept "jupyter-demo") (Concept "demo"))')
    print("Linked:", r.get("result", r))

    # Count
    r = await ask('(cog-count-atoms)')
    print("Total atoms:", r.get("result", r))

asyncio.run(main())
````
